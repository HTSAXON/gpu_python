```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from hummingbird.ml import convert, load
```

```python
records = 5*100000
```

```python
# Create some random data for binary classification
num_classes = 2
X = np.random.rand(records, 28)
y = np.random.randint(num_classes, size=records)
```

```python
# Create and train a model (scikit-learn RandomForestClassifier in this case)
skl_model = RandomForestClassifier(n_estimators=20, max_depth=10, n_jobs=-1)
```

```python
skl_model.fit(X, y)
```

```python
# Use Hummingbird to convert the model to PyTorch
model = convert(skl_model, 'pytorch')
```

```python
# Run predictions on CPU
model.predict(X)
```

```python
# Run predictions on GPU
model.to('cuda')
model.predict(X)
```

```python
# Save the model
model.save('hb_model')
```

```python
# Load the model back
model = load('hb_model')
```
